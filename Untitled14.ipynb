## Data loading

### Subtask:
Load the provided CSV files into pandas DataFrames.


**Reasoning**:
Import pandas and load the provided CSV files into pandas DataFrames. Also, read the data description file.



```python
import pandas as pd

# Load the CSV files into pandas DataFrames
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
df_submission = pd.read_csv('sample_submission.csv')

# Read the data description file
with open('data_description.txt', 'r') as file:
    data_description = file.read()

# Display the first few rows of each DataFrame for a quick overview
display(df_train.head())
display(df_test.head())
display(df_submission.head())
print(data_description)
```

## Data exploration

### Subtask:
Explore the loaded datasets to understand their structure, data types, missing values, and basic statistics.


**Reasoning**:
Analyze the dataframes using .info(), .describe(), and .value_counts() to understand their structure, data types, missing values, and basic statistics.



```python
# Display info for df_train and df_test
display(df_train.info())
display(df_test.info())

# Display descriptive statistics for numerical features in df_train and df_test
display(df_train.describe())
display(df_test.describe())

# Analyze categorical features in df_train and df_test
categorical_cols_train = df_train.select_dtypes(include=['object']).columns
categorical_cols_test = df_test.select_dtypes(include=['object']).columns

for col in categorical_cols_train:
    print(f"\nValue counts for {col} in df_train:")
    display(df_train[col].value_counts())

for col in categorical_cols_test:
    print(f"\nValue counts for {col} in df_test:")
    display(df_test[col].value_counts())
```

## Data visualization

### Subtask:
Visualize the distribution of numerical features and identify potential outliers using histograms and box plots.


**Reasoning**:
Visualize the distribution of numerical features using histograms and box plots, identify potential outliers, and save the plots. Document observations about the distributions.



```python
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Create a directory to store the plots if it doesn't exist
if not os.path.exists('plots'):
    os.makedirs('plots')

numerical_cols_train = df_train.select_dtypes(include=['int64', 'float64']).columns
numerical_cols_test = df_test.select_dtypes(include=['int64', 'float64']).columns

for col in numerical_cols_train:
    if col != 'Id' and col != 'SalePrice': # Skip 'Id' and 'SalePrice' for histogram
        plt.figure(figsize=(8, 6))
        sns.histplot(df_train[col], kde=True)
        plt.title(f'Histogram of {col} (Training Data)')
        plt.xlabel(col)
        plt.ylabel('Frequency')
        plt.savefig(os.path.join('plots', f'histogram_{col}_train.png'))
        plt.close() # Close the plot to avoid overlapping

    plt.figure(figsize=(8, 6))
    sns.boxplot(y=df_train[col])
    plt.title(f'Boxplot of {col} (Training Data)')
    plt.ylabel(col)
    plt.savefig(os.path.join('plots', f'boxplot_{col}_train.png'))
    plt.close() # Close the plot

for col in numerical_cols_test:
    if col != 'Id': #Skip 'Id' for histogram in test data
        plt.figure(figsize=(8, 6))
        sns.histplot(df_test[col], kde=True)
        plt.title(f'Histogram of {col} (Test Data)')
        plt.xlabel(col)
        plt.ylabel('Frequency')
        plt.savefig(os.path.join('plots', f'histogram_{col}_test.png'))
        plt.close() # Close the plot

    plt.figure(figsize=(8, 6))
    sns.boxplot(y=df_test[col])
    plt.title(f'Boxplot of {col} (Test Data)')
    plt.ylabel(col)
    plt.savefig(os.path.join('plots', f'boxplot_{col}_test.png'))
    plt.close() # Close the plot

print("Histograms and boxplots saved to 'plots' directory.")
```
