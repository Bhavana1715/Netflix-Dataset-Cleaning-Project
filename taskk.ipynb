```python
import pandas as pd
```

```python
#Loading the dataset
df=pd.read_csv('/content/drive/MyDrive/netflix_titles.csv')
df.head(10)
```

```python
#Dataset exploring
print("\n Shape Of the dataframe")
print(df.shape)
print("\n Data types of the columns")
display(df.dtypes)
#Dataset Information
print("\n Dataset Information")
print(df.info())
```

```python
#glance of numerical data
print("\n Glance of numerical data")
print(df.describe())
#glance of categorical data
print("\n Glance of categorical data")
for col in df.select_dtypes(include=['object']).columns:
    print(f"\nColumn: {col}")
    print(f"Unique values: {df[col].unique()[:20]}") # Limiting to first 20 for brevity
    print(f"Value counts:\n{df[col].value_counts()[:10]}")

```

```python
#Missing Values
missing_values=df.isnull().sum()
missing_percentage=(missing_values/len(df))*100
missing_info=pd.DataFrame({"Missing Values":missing_values,"Missing_percentage":missing_percentage})
display(missing_info)
#Duplicate Rows
duplicate_rows = df[df.duplicated()]
num_duplicates = len(duplicate_rows)
print(f"\nNumber of duplicate rows: {num_duplicates}")
print("\nDuplicate Rows:")
print(duplicate_rows)
display(df.columns)

```

```python
# 1. Handle missing values correctly
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')
df['country'] = df['country'].fillna('Unknown')

# 2. Smart filling for date
movie_common_date = df[df['type'] == 'Movie']['date_added'].mode()[0]
tvshow_common_date = df[df['type'] == 'TV Show']['date_added'].mode()[0]

def smart_fill(row):
    if pd.isnull(row['date_added']):
        if row['type'] == 'Movie':
            return movie_common_date
        else:
            return tvshow_common_date
    else:
        return row['date_added']

df['date_added'] = df.apply(smart_fill, axis=1)

# 3. Fill Rating and Duration
df['rating'] = df['rating'].fillna('Not Rated')

if not df['duration'].mode().empty:
    df['duration'] = df['duration'].fillna(df['duration'].mode()[0])
else:
    df['duration'] = df['duration'].fillna('Unknown')

# 4. Data type conversions
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['release_year'] = df['release_year'].astype('int64')
df.head()

```
